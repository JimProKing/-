```python
import google.generativeai as genai
from IPython.display import display, clear_output
import time

# --- 1. 텍스트 파일 읽기 (CP949) ---
def read_txt_file(file_path):
    with open(file_path, "r", encoding="cp949") as f:
        return f.read()

# --- 2. 긴 텍스트를 자동으로 청크 분할 ---
def split_text_auto(text, max_chunk_chars=500):
    """
    긴 텍스트를 max_chunk_chars 단위로 분할
    """
    return [text[i:i+max_chunk_chars] for i in range(0, len(text), max_chunk_chars)]

# --- 3. Gemini Pro 모델로 분석 (청크 단위 안전 호출) ---
def analyze_txt_safely(txt_content, model_name="gemini-1.5-pro", max_output_tokens=200, max_chunk_chars=500):
    API_KEY = "AIzaSyBIcCHStQ3-EAcfAmdZcQLZaEDiVLlHrT0"  # 여기에 본인 키 입력
    genai.configure(api_key=API_KEY)
    
    from google.generativeai import types
    model = genai.GenerativeModel(model_name)
    generation_config = types.GenerationConfig(max_output_tokens=max_output_tokens)

    # 텍스트 청크 분할
    chunks = split_text_auto(txt_content, max_chunk_chars=max_chunk_chars)

    full_response = ""
    total_prompt_tokens = 0
    total_output_tokens = 0

    for i, chunk in enumerate(chunks, start=1):
        prompt = f"아래는 2025 1분기 재무상태표(은행) 데이터 청크 {i}입니다. 주요 재무사항과 특징을 요약해줘:\n\n{chunk}"
        try:
            stream = model.generate_content(prompt, generation_config=generation_config, stream=True)
        except Exception as e:
            print(f"Chunk {i} 분석 중 오류 발생: {e}")
            continue

        chunk_response = ""
        for c in stream:
            if c.text:
                chunk_response += c.text
                clear_output(wait=True)
                display(full_response + chunk_response)
                time.sleep(0.05)

        # 사용량 누적
        usage = stream.usage_metadata
        total_prompt_tokens += usage.prompt_token_count
        total_output_tokens += usage.candidates_token_count

        full_response += f"\n--- Chunk {i} Analysis ---\n{chunk_response}\n"

        # 서버 과부하 방지
        time.sleep(1)

    # 비용 계산 (Pro 모델 기준)
    rate_in, rate_out = 1.25, 5.00
    usd_cost = (total_prompt_tokens * rate_in + total_output_tokens * rate_out) / 1_000_000
    krw_cost = usd_cost * 1393

    print("\n--- 총 사용량 ---")
    print(f"프롬프트 토큰: {total_prompt_tokens}")
    print(f"출력 토큰: {total_output_tokens}")
    print(f"총 토큰: {total_prompt_tokens + total_output_tokens}")
    print(f"\n--- 예상 비용 ---\n달러: ${usd_cost:.6f} / 원화: 약 ₩{krw_cost:.2f}")

    return full_response

# --- 4. 실행 ---
txt_content = read_txt_file("./2025_1분기보고서_01_재무상태표_은행_20250805.txt")
analysis_result = analyze_txt_safely(txt_content, model_name="gemini-1.5-pro", max_output_tokens=200, max_chunk_chars=500)
```
